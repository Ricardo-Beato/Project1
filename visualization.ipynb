```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', None)
```

#### 1 Crossing column Country(Australia) with Age to check the age distribution of Australians attacked by sharks

```python
sharks_q1 = pd.read_csv("C:/Users/mocid/Ironhack/labs/projects/Project1/data/Q1 attacks.csv",encoding='latin1')
sharks_q1.sample(50)
```

```python
sharks_q1_australia = sharks_q1[sharks_q1["Country"] == "AUSTRALIA"]
sharks_q1_australia
```

```python
sns.set(style="darkgrid")
custom_palette = ["#00539C", "#2972D1", "#5699E1", "#80B6F4", "#A5CBFA"] #I want blue colours since... it's sharks and the sea. Also fun fact: I'm colourblind.
sns.set_palette(custom_palette)
```

```python
#using a histogram:

sharks_q1_australia_hist = sns.histplot(x=sharks_q1_australia.Age); 

for p in sharks_q1_australia_hist.patches:
    sharks_q1_australia_hist.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                textcoords='offset points')
    
sharks_q1_australia_hist.figure.savefig("./images/sharks_q1_australia_hist.png", dpi=1000)
```

#### 3 Crossing column Type[Provoked] with country to check what countries incite more shark attacks

```python
sharks_q3 = pd.read_csv("C:/Users/mocid/Ironhack/labs/projects/Project1/data/Q3 attacks.csv",encoding='latin1')
sharks_q3.sample(50)
```

```python
sharks_q3.groupby("Country")["Type"].count()
```

```python
provoked_count = sharks_q3["Country"].value_counts()
provoked_count
```

```python
other_types_count = sharks_q3[sharks_q3["Type"] != "Provoked"]["Country"].value_counts()
other_types_count
```

```python
relative_count = (provoked_count / (provoked_count + other_types_count)).sort_values(ascending=False)
relative_count
```

```python
# the absolute count chart:
plt.figure(figsize=(16, 9))
sns.set(style="darkgrid")
sns.set_palette(custom_palette)
absolute_count_of_provoked = sns.countplot(x="Country", data=sharks_q3, order=provoked_count.index[:5])
plt.title("Top 5 Countries with the Most 'Provoked' Shark Incidents (Absolute Count)")
plt.xticks(rotation=45)

# Annotate the bars with their counts
for i, p in enumerate(absolute_count_of_provoked.patches):
    absolute_count_of_provoked.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                textcoords='offset points')
    p.set_facecolor(custom_palette[i])  # Set a different shade of blue for each bar

plt.show()

absolute_count_of_provoked.figure.savefig("./images/q3absolute_count_of_provoked.png", dpi=1000)
```

```python
# the relative count of povoked chart:
plt.figure(figsize=(20, 12))
sns.set(style="darkgrid")
sns.set_palette(custom_palette)
relative_count_of_provoked = sns.barplot(x=relative_count.index[:5], y=relative_count[:5])
plt.title("Top 5 Countries with the Highest Relative Count of 'Provoked' Shark Incidents")
plt.xticks(rotation=45)

# Annotate the bars with their values
for i, p in enumerate(relative_count_of_provoked.patches):
    relative_count_of_provoked.annotate(f'{p.get_height():.2%}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                textcoords='offset points')
    p.set_facecolor(custom_palette[i])  # Set a different shade of blue for each bar
    
plt.margins(x=0.1) #the labels were being cut

plt.show()

relative_count_of_provoked.figure.savefig("./images/q3relative_count_of_provoked.png", dpi=1000)
```

#### 4 When were women the most targeted in USA between 1995-2018?

```python
sharks_q4 = pd.read_csv("C:/Users/mocid/Ironhack/labs/projects/Project1/data/Q4 attacks.csv",encoding='latin1')
sharks_q4.sample(50)
```

```python
condition_1_sharks_q4 = sharks_q4["Year"].between(1995, 2018)
condition_2_sharks_q4 = sharks_q4["Sex"] == "F"
condition_3_sharks_q4 = sharks_q4["Country"] == "USA"


sharks_q4_women_per_years = sharks_q4[condition_1_sharks_q4 & condition_2_sharks_q4 & condition_3_sharks_q4]
sharks_q4_women_per_years.sample(50)
```

```python
sns.set(style="darkgrid") #do I need to do this everytime?
custom_palette = ["#00539C", "#2972D1", "#5699E1", "#80B6F4", "#A5CBFA"] #I want them blues again
sns.set_palette(custom_palette)

plt.figure(figsize=(12, 6))
q4_women_under_attack_USA = sns.countplot(x="Year", data=sharks_q4_women_per_years, palette=custom_palette)

for p in q4_women_under_attack_USA.patches:
    q4_women_under_attack_USA.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                textcoords='offset points')

plt.title("Women under attack in the USA (1995-2018)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)  # the labels in the x-axis where overlapping like crazy

plt.show()

q4_women_under_attack_USA.figure.savefig("./images/q4_women_under_attack_USA.png", dpi=1000)
```

```python
#well, what about men?
condition_4_sharks_q4 = sharks_q4["Year"].between(1995, 2018)
condition_5_sharks_q4 = sharks_q4["Sex"] == "M"
condition_6_sharks_q4 = sharks_q4["Country"] == "USA"


sharks_q4_men_per_years = sharks_q4[condition_4_sharks_q4 & condition_5_sharks_q4 & condition_6_sharks_q4]
sharks_q4_men_per_years.sample(50)
```

```python
sns.set(style="darkgrid") #do I need to do this everytime?
custom_palette = ["#00539C", "#2972D1", "#5699E1", "#80B6F4", "#A5CBFA"] #I want them blues again
sns.set_palette(custom_palette)

plt.figure(figsize=(12, 6))
q4_men_under_attack_USA = sns.countplot(x="Year", data=sharks_q4_men_per_years, palette=custom_palette)

for p in q4_men_under_attack_USA.patches:
    q4_men_under_attack_USA.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                textcoords='offset points')

plt.title("Men under attack in the USA (1995-2018)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)  # the labels in the x-axis where overlapping like crazy

plt.show()

q4_men_under_attack_USA.figure.savefig("./images/q4_men_under_attack_USA.png", dpi=1000)
```
