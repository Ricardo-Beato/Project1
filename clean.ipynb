```python
"""
I will define different "categories" for the time:
(reference: https://www.britannica.com/dictionary/eb/qa/parts-of-the-day-early-morning-late-morning-etc)

Morning     5 am to 12 pm (noon)
Afternoon     12 pm to 5 pm
Evening     5 pm to 9 pm
Night         9 pm to 4 am

"""

```

```python
#first off I'll deal with the data that's got an entry in the proper form "[Hour]h[minute]". 
#I'm actually creating a new column to categorize these times:

import re
#importing regular exprssns

time_categories = ["Morning", "Afternoon", "Evening", "Night"]
time_categories_intervals = [5, 12, 17, 21, 24] 

#as we've seen in class, to apply a function over a column we need to use the apply method.
#but first I'll create the function itself:

# Function to categorize the time intervals. Once again - had to ask for help online as Regex and I are not friends yet.
def categorize_the_time(time_str):
    """
categorizing the time into the 4 different categories I've defined ("Morning", "Afternoon", "Evening", "Night").
I realized lots of entries are done in the format HHhMM so I'm targeting these entries by capturing 2 groups (hour and minute)

r'(\d{1,2})h(\d{2})
is precisely tackling these 2 groups. 
\d - to match any numeric digit (0-9)
{1,2} - means the numbers can vary between on (imagine 9am) and 2 (imagine 11am)
h - literally people wrote "h" in the time entry
(\d{2}) - in the second group there are always gonna be 2 digits, even if it's 8h02 the 2 is actually "02"


"""    
    
    # Extract hours and minutes using regular expressions
    match = re.match(r'(\d{1,2})h(\d{2})', time_str)    
    if match:
        hour = int(match.group(1))
        minute = int(match.group(2))
        # Convert to a single hour value (0-23)
        hour_value = hour + minute / 60

        # Categorize based on time ranges
        if 5 <= hour_value < 12:
            return "Morning"
        elif 12 <= hour_value < 17:
            return "Afternoon"
        elif 17 <= hour_value < 21:
            return "Evening"
        else:
            return "Night"
    else:
        return "Unknown"

# Create a new column 'Time of Day' based on 'Time' column
#IGNORE THIS CODE: sharks['Time of Day'] = sharks['Time'].apply(categorize_the_time)
sharks_q2['Time of Day'] = sharks_q2['Time'].apply(lambda time_str: categorize_the_time(str(time_str))) 
#I was getting an error with the code above (the IGNORE one) as I was not casting the newly created col into a string. 

```
